```python
import pandas as pd
df = pd.read_csv('cleaned_processed.cleveland.data', header = None)
df.head()
```

```python

data = df.drop(df.columns[[13]],axis =1)
label = df[df.columns[[13]]]

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import tree 
import numpy as np
import pydotplus

dataTrain, dataTest, labelTrain,labelTest = train_test_split(data,label, test_size = 0.2)

def dtree_grid_search(data, target, nfolds):
    param_grid = {'criterion':['gini','entropy'], 'max_depth':np.arange(2,10)}
    d_tree = tree.DecisionTreeClassifier()
    
    gridsearch = GridSearchCV(d_tree, param_grid, cv=nfolds)
    gridsearch.fit(data, target)
    return gridsearch.best_params_
print("Best hyperparameter for this case:",dtree_grid_search(data, label, 5))


decision_tree_entropy = tree.DecisionTreeClassifier(random_state=0, max_depth=3, criterion = "gini")
decision_tree_entropy.fit(dataTrain,labelTrain)

result_predict = decision_tree_entropy.predict(dataTest)
accuracy = decision_tree_entropy.score(dataTest, labelTest)

print('Model A on Gini')
print('Accuracy of decision tree is %.5f' %(accuracy))

from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
# https://pydotplus.readthedocs.io/ for generating generates a GraphViz representation of the decision tree.

dot_data = StringIO()
export_graphviz(decision_tree_entropy, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
